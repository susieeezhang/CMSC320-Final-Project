# Functions to Find the Best Match for a Given Participant

```python
# Define a function to calculate compatibility score based on the specified questions
def calculate_compatibility_score_specific(participant1, participant2):
    score = 0
    matched_questions = []

    for question in questions:
        if participant1[question] == participant2[question]:
            score += 1
            matched_questions.append(question)

    return score, matched_questions


def calculate_compatibility_score_with_preference(participant1, participant2):
    if participant1['WantToDate'] != participant2['Gender']:
        return 0, []  # No compatibility if gender preference doesn't match

    score = 0
    matched_questions = []

    for question in questions:
        if participant1[question] == participant2[question]:
            score += 1
            matched_questions.append(question)

    return score, matched_questions




# Define a function to calculate compatibility score and return all matches with highest score
def find_best_matches_with_tie_breaking(participant_id, participants):
    highest_score = float('-inf')
    best_matches = []

    for other_id, other_participant in participants.items():
        if participant_id != other_id:
            score, matched_questions = calculate_compatibility_score_with_preference(participants[participant_id], other_participant)
            if score > highest_score:
                highest_score = score
                best_matches = [(other_id, matched_questions)]
            elif score == highest_score:
                best_matches.append((other_id, matched_questions))

    return best_matches, highest_score

# Create dictionary of participants
participants = {row['ID']: row for index, row in matching_data_en.iterrows()}

# List to store matches
matches = []
matched_ids = set()
```

# Matching Process

```python

# Create a dictionary of participants for easy access
participants = {row['ID']: row for index, row in matching_data_en.iterrows()}

# List to store the matches
matches = []

# List to keep track of matched participants
matched_ids = set()

# Matching process
for participant_id, participant in participants.items():
    best_matches, score = find_best_matches_with_tie_breaking(participant_id, participants)

    if best_matches:
        # Store the match
        match_info = {
            'Participant ID': participant_id,
            'Participant Name': participant['Name'],
            'Best Matches': [],
            'Score': score
        }
        for best_match_id, matched_questions in best_matches:
            match_info['Best Matches'].append({
                'Match ID': best_match_id,
                'Match Name': participants[best_match_id]['Name'],
                'Matched Questions': [q.split('.')[0] for q in matched_questions]
            })
            # Add to matched IDs
            matched_ids.add(best_match_id)

        matches.append(match_info)


# Create a result table
result_table = []

for match in matches:
    for best_match in match['Best Matches']:
        matching_percentage = (match['Score'] / total_questions) * 100
        result_table.append({
            'Participant ID': match['Participant ID'],
            'Participant Name': match['Participant Name'],
            'Best Match ID': best_match['Match ID'],
            'Best Match Name': best_match['Match Name'],
            '"恋爱观问题" Matching Percentage': f"{matching_percentage:.2f}%",
            'Number of Matched Questions': match['Score'],
            'Matched Questions': ', '.join(best_match['Matched Questions'])
        })

# Convert result table to a DataFrame for better display
pd.set_option('display.max_rows', 500)

result_df = pd.DataFrame(result_table)
result_df
```
